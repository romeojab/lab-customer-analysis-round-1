```python
import pandas as pd
file1 = pd.read_csv('lab-customer-analysis-round-1/files_for_lab/csv_files/file1.csv')
file2 = pd.read_csv('lab-customer-analysis-round-1/files_for_lab/csv_files/file2.csv')
file3 = pd.read_csv('lab-customer-analysis-round-1/files_for_lab/csv_files/file3.csv')
```

```python
file1 = file1.rename(columns={'ST': 'State', 'GENDER':'Gender'})
file2 = file2.rename(columns={'ST': 'State', 'GENDER':'Gender'})
file3 = file3.rename(columns={'ST': 'State', 'GENDER':'Gender'})
```

```python
file2 = file2[['Customer', 'State', 'Gender', 'Education', 'Customer Lifetime Value',
       'Income', 'Monthly Premium Auto', 'Number of Open Complaints',
       'Policy Type', 'Vehicle Class', 'Total Claim Amount']]

file3 = file3[['Customer', 'State', 'Gender', 'Education', 'Customer Lifetime Value',
       'Income', 'Monthly Premium Auto', 'Number of Open Complaints',
       'Policy Type', 'Vehicle Class', 'Total Claim Amount']]
```

```python
list1 = file1['Customer Lifetime Value']
listfile1 = []
test = []

for value in list1:
    listfile1.append(str((value)))
    
for value in listfile1:
    test.append(value.replace('%',''))

test = pd.to_numeric(test, errors='coerce')
                              
file1['Customer Lifetime Value'] = test/100

file1
```

```python
list2 = file2['Customer Lifetime Value']
listfile2 = []
test2 = []

for value in list2:
    listfile2.append(str((value)))
    
for value in listfile2:
    test2.append(value.replace('%',''))

test2 = pd.to_numeric(test2, errors='coerce')
                              
file2['Customer Lifetime Value'] = test2/100

file2
```

```python
data = pd.concat([file1, file2, file3], axis=0)
data.shape
data
```
